```python
#Step 1: Data Cleaning:
# Based on above output we can specifically observe that some variables need data cleaning:
# they are: Stadium Type, Turf, GameWeather, WindSpeed, WindDirection

#Tackling them one by one: 
# Data Cleaning:StadiumType

nfl_train['StadiumType'].unique()

#StadiumType: First changing case to lower:
nfl_train['StadiumType'] = nfl_train['StadiumType'].str.lower()
nfl_train[nfl_train['StadiumType'].isin (['cloudy','outdoors','ourdoor','outside','oudoor','open','outddors','outdor'])] = 'outdoor'  
nfl_train[nfl_train['StadiumType'].isin (['dome, closed','domed, closed'])] = 'closed dome'  #nfl_train[nfl_train["StadiumType"] .isin (['retr. roof - open','outdoor retr roof-open'])] = 'retr. roof-open' 
nfl_train[nfl_train['StadiumType'].isin (['retr. roof - closed','retr. roof closed','retr. roof-closed', 'retr. roof-open'])] = 'retractable roof'
nfl_train[nfl_train['StadiumType'].isin (['outdoor retr roof-open','retr. roof - open'])] = 'retractable roof' 
nfl_train[nfl_train['StadiumType'].isin (['domed, open','domed','closed dome'])] = 'dome' 
nfl_train[nfl_train['StadiumType'].isin (['indoors','indoor, open roof','indoor, roof closed'])] = 'indoor'

nfl_train.StadiumType.unique()
#nfl_train.StadiumType.value_counts(dropna = False)

nfl_train.head(5)
```
